```python
import numpy as np
import matplotlib.colors
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import style

ft = int(input('Input count of features:'))
ob  = int(input('Input count of objects:'))
mass = []

#  
seed = 12
rand = np.random.RandomState(seed)

#  
dist_list = ['uniform','normal','exponential','lognormal','chisquare','beta']
count_dist_list = len(dist_list)
param_list = ['-1,1','0,1','1','0,1','2','0.5,0.9']
colors_list = ['green','blue','yellow','cyan','magenta','pink']

#  
n_cols = 3
fig, ax = plt.subplots(nrows=ft // n_cols + 1, ncols=n_cols,figsize=(12,7))
plt_ind_list = [[i, j] for i in range(ft // n_cols + 1) for j in range(3)]

#    ()
for i in range(ft):
    #    
    num = rand.randint(1, count_dist_list) - 1
    #   
    x = eval('rand.' + dist_list[num] + '(' + param_list[num] + ',' + str(ob) + ')')
    a, b = plt_ind_list[i]
    ax[a, b].hist(x, bins=50, color=colors_list[num])
    ax[a, b].set_title(dist_list[num])
    mass.append(x)
    
fig.subplots_adjust(hspace=0.4,wspace=.3) 
plt.suptitle('Sampling from Various Distributions',fontsize=20)
plt.show()
#print(mass)
```
